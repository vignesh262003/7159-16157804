```python
# STEP 1: Import packages
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader

# STEP 2: Define CNN + GRU model (same as before)
class CNN_GRU_Model(nn.Module):
    def __init__(self, num_classes=10):
        super(CNN_GRU_Model, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, padding=1)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.gru = nn.GRU(input_size=64*8, hidden_size=128, batch_first=True)
        self.fc = nn.Linear(128, num_classes)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool1(x)
        x = F.relu(self.conv2(x))
        x = self.pool2(x)
        x = x.permute(0, 2, 1, 3)
        x = x.reshape(x.size(0), x.size(1), -1)
        out, _ = self.gru(x)
        out = out[:, -1, :]
        return self.fc(out)

# STEP 3: Load CIFAR-10 dataset
transform = transforms.Compose([transforms.ToTensor()])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
trainloader = DataLoader(trainset, batch_size=64, shuffle=True)

testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = DataLoader(testset, batch_size=64, shuffle=False)

# STEP 4: Training utilities
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = CNN_GRU_Model().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# STEP 5: Training loop
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    epoch_loss = running_loss / len(trainloader.dataset)
    epoch_acc = 100 * correct / total

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.2f}%")
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# CIFAR-10 classes
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck']

# Load test set
transform = transforms.Compose([transforms.ToTensor()])
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=1, shuffle=False)

# Set model to evaluation mode
model.eval()

# Collect misclassified samples
misclassified = []

with torch.no_grad():
    for images, labels in testloader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)

        if predicted != labels:
            misclassified.append((images.cpu(), labels.item(), predicted.item()))

        if len(misclassified) >= 10:
            break  # Show only first 10 misclassifications

# Plot misclassified images
plt.figure(figsize=(12, 6))
for idx, (img, true_label, pred_label) in enumerate(misclassified):
    plt.subplot(2, 5, idx+1)
    img = img.squeeze().permute(1, 2, 0).numpy()
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    plt.title(f"True: {classes[true_label]}\nPred: {classes[pred_label]}", color='red')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# CIFAR-10 classes
classes = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer',
           'Dog', 'Frog', 'Horse', 'Ship', 'Truck']

# Load test set
transform = transforms.Compose([transforms.ToTensor()])
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)

# Set model to eval mode
model.eval()

# Get predictions and true labels
all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in testloader:
        images = images.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Compute confusion matrix
cm = confusion_matrix(all_labels, all_preds)

# Plot the confusion matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np

# CIFAR-10 class names
classes = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer',
           'Dog', 'Frog', 'Horse', 'Ship', 'Truck']

# Load test data
transform = transforms.Compose([transforms.ToTensor()])
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=1, shuffle=False)

# Put model in evaluation mode
model.eval()

# Collect ambiguous predictions (confidence < 0.5)
ambiguous = []

with torch.no_grad():
    for images, labels in testloader:
        images = images.to(device)
        outputs = model(images)
        probs = F.softmax(outputs, dim=1)
        conf, predicted = torch.max(probs, 1)

        if conf.item() < 0.5:
            ambiguous.append((images.cpu().squeeze(), labels.item(), predicted.item(), conf.item()))
        if len(ambiguous) >= 25:
            break  # limit to 25 examples

# Plot ambiguous images in grid
plt.figure(figsize=(12, 8))
for idx, (img, true_label, pred_label, confidence) in enumerate(ambiguous):
    plt.subplot(5, 5, idx + 1)
    img = img.permute(1, 2, 0).numpy()
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    title = f"T:{classes[true_label]}\nP:{classes[pred_label]}\n({confidence:.2f})"
    plt.title(title, fontsize=8)
    plt.axis('off')

plt.suptitle("Ambiguous Predictions (Softmax < 0.5)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np

# CIFAR-10 class names
classes = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer',
           'Dog', 'Frog', 'Horse', 'Ship', 'Truck']

# Set model to evaluation mode
model.eval()

ambiguous_wrong = []

# Threshold for ambiguity
threshold = 0.5

# Collect ambiguous + incorrect predictions
with torch.no_grad():
    for images, labels in testloader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        probs = F.softmax(outputs, dim=1)
        max_probs, preds = torch.max(probs, dim=1)

        for img, prob, pred, true in zip(images, max_probs, preds, labels):
            if prob.item() < threshold and pred.item() != true.item():
                ambiguous_wrong.append((img.cpu(), true.item(), pred.item(), prob.item()))

            if len(ambiguous_wrong) >= 25:
                break
        if len(ambiguous_wrong) >= 25:
            break

# Plot ambiguous + incorrect predictions
fig, axs = plt.subplots(5, 5, figsize=(15, 10))
fig.suptitle("Ambiguous + Incorrect Predictions", fontsize=16)

for i, (img, true_label, pred_label, confidence) in enumerate(ambiguous_wrong):
    ax = axs[i // 5][i % 5]
    img = img.permute(1, 2, 0).numpy()
    img = np.clip(img, 0, 1)
    ax.imshow(img)
    ax.set_title(f"T:{classes[true_label]}\nP:{classes[pred_label]}", fontsize=10)
    ax.axis('off')

plt.tight_layout()
plt.subplots_adjust(top=0.88)  # for title
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Reshape, Conv2D, BatchNormalization,
                                     MaxPooling2D, LSTM, Dropout, Dense)

# Build the model
model = Sequential([
    Reshape((32, 32, 3), input_shape=(32*32*3,)),                      # (None, 32, 32, 3)
    Conv2D(32, (3, 3), padding='same', activation='relu'),            # (None, 32, 32, 32)
    BatchNormalization(),                                             # (None, 32, 32, 32)
    MaxPooling2D(pool_size=(2, 2)),                                   # (None, 16, 16, 32)

    Conv2D(64, (3, 3), padding='same', activation='relu'),            # (None, 16, 16, 64)
    BatchNormalization(),                                             # (None, 16, 16, 64)
    MaxPooling2D(pool_size=(2, 2)),                                   # (None, 8, 8, 64)

    Reshape((8, 512)),                                                # (None, 8, 512)
    LSTM(128),                                                        # (None, 128)
    Dropout(0.5),                                                     # (None, 128)
    Dense(10, activation='softmax')                                   # (None, 10)
])

# Print model summary
model.summary()
```
